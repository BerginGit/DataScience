# Codigo

```python
url = "https://www.fab.mil.br/ingresso/arquivos/2022/cpcar/cpcar2023_resultado_provisorio.pdf"
```

```python
pages_count = 127
```

```python
aprovados_pages_list = list(range(1,pages_count)) # apenas as paginas dos aprovados
```

```python
pages_count = 1218
```

```python
reprovados_pages_list = list(range(127,pages_count)) # apenas as paginas dos aprovados
```

```python
# ler o pdf de uma url

pdf_list = tabula.read_pdf(url,pages= aprovados_pages_list, stream=True)
```

```python
candidatos_pdf_list = []
```

```python
candidatos_pdf_list.append(pdf_list.copy())
```

```python
# ler o pdf de uma url

pdf_list = tabula.read_pdf(url,pages= reprovados_pages_list, stream=True)
```

```python
candidatos_pdf_list.append(pdf_list.copy())
```

```python
candidatos_pdf_list[0][:2]
```

```python
candidatos_pdf_list[1][:2]
```

```python
candidatos_pandas_df = []
```

```python
candidatos_pandas_df.append(pd.concat(candidatos_pdf_list[0], ignore_index=True)) #juntar tudo em um dataframe
```

```python
candidatos_pandas_df.append(pd.concat(candidatos_pdf_list[1], ignore_index=True)) #juntar tudo em um dataframe
```

```python
candidatos_pandas_df[0][:10]
```

```python
candidatos_pandas_df[1][:10]
```

Descartando colunas vazias ou nao uteis e depois dividir a coluna Class Port em duas

```python
candidatos_pandas_df[0] = candidatos_pandas_df[0].drop('Nome', axis=1)
```

```python
candidatos_pandas_df[0] = candidatos_pandas_df[0].drop('Unnamed: 0', axis=1)
```

```python
candidatos_pandas_df[0] = candidatos_pandas_df[0].drop('Nome Inscrição', axis=1)
```

```python
candidatos_pandas_df[1] = candidatos_pandas_df[1].drop('Nome', axis=1)
```

```python
candidatos_pandas_df[1] = candidatos_pandas_df[1].drop('Unnamed: 0', axis=1)
```

```python
candidatos_pandas_df[0][:10]
```

```python
candidatos_pandas_df[1][:10]
```

```python
candidatos_pandas_df[0].rename(columns={'Class Port': 'Class','Unnamed: 1': 'Port'}, inplace=True) #renomeando algumas colunas
```

```python
candidatos_pandas_df[0].Port = candidatos_pandas_df[0].Class #passando os valores apenas para nao ficar com NaN e futuramente tratar
```

```python
candidatos_pandas_df[0].Port[:10]
```

```python
df = []
```

```python
df.append(candidatos_pandas_df[0].dropna(axis=0)) #remocao dos valores NaN
```

```python
df.append(candidatos_pandas_df[1].dropna(axis=0)) #remocao dos valores NaN
```

```python
df[0].reset_index(drop=True, inplace=True) #reiniciando os indices apos remocao de linhas
```

```python
df[1].reset_index(drop=True, inplace=True) #reiniciando os indices apos remocao de linhas
```

```python
df[0][:10]
```

```python
df[1][:10]
```

```python
df[0][['Class', 'Port']] = df[0]['Class'].str.split(' ', 1, expand=True) #usando split para dividir a coluna Class que tinha os valores de Class e Port juntos para seus devidos lugares
```

```python
df[0]["Inscrição"] = df[0]["Inscrição"].apply(np.int32)
```

```python
df[1]["Inscrição"] = df[1]["Inscrição"].apply(np.int32)
```

```python
df[0][:10]
```

```python
df[1][:10]
```

```python
df[0].to_csv("aprovados_EPCAR.csv", encoding = "utf-8" , index=False)
```

```python
df[1].to_csv("reprovados_EPCAR.csv", encoding = "utf-8" , index=False)
```
